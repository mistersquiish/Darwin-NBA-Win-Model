# Exploratory Analysis of Data

To analyze this problem, we decided to use regular season NBA data from the past 10 years. We scraped this data from the official NBA website (https://stats.nba.com/teams/), and added several of our own columns as well. The official NBA data had dozens of categories, including Traditional, Advanced, Four Factors, Misc, Scoring, Opponent, and Defense; each of these categories contained different pieces of data. For example, true shooting percentage (a statistic that measures how efficiently a player shot the ball) fell under Advanced, but field goal percentage (the number of shots made over the number of shots taken) fell under Traditional.

Our feature engineered columns were number players who received NBA awards during that season. These awards are given to top players at their position as voted by fans, media, and coaches. The features we created were:
- Number of All Star players
- Number of All-NBA 1st team players
- Number of All-NBA 2nd team players
- Number of All-NBA 3rd team players

Before running Darwin, we decided to explore our full dataset and its features.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("NBA_Combined_Data.csv")
df.head()
```

First, we decided to plot Win% against Points to see if there was any correlation. We found that there was generally a positive correlation, so the more points a team scores, the higher its winning percentage is likely to be.

Then, we decided plot Win% against Defensive Rating. We were very surprised by the results; essentially, being a good defensive team correlates to a lower winning %.

```python
plt.scatter(df['Label'], df['DEF RTG'])
plt.xlabel('Win %')
plt.ylabel('Defensive Rating')
plt.show()
```

We continued to plot Win% against variables to check correlations and explore the data.

```python
plt.scatter(df['Label'], df['TOV'])
plt.xlabel('Win%')
plt.ylabel('Turnover ')
plt.show()
```

## Darwin Model v2.0

The section below illustrates the steps to setup the Darwin environment, upload our datsets, clean the datsets, create our model, and test our model. 

### Set Darwin SDK

```python
from amb_sdk.sdk import DarwinSdk
ds = DarwinSdk()
ds.set_url('https://amb-demo-api.sparkcognition.com/v1/')
```

### Environment Variables

```python
# Set path to datasets - The default below assumes Jupyter was started from amb-sdk/examples/Enterprise/
# Modify accordingly if you wish to use your own data
PATH_TO_DATASET='sets/'
TRAIN_DATASET='nba_train.csv'
TEST_DATASET='nba_test.csv'
# A timestamp is used to create a unique name in the event you execute the workflow multiple times or with 
# different datasets.  File names must be unique in Darwin.
import datetime
ts = '{:%Y%m%d%H%M%S}'.format(datetime.datetime.now())
```

### Import necessary libraries

```python
# Import necessary libraries
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import Image
from time import sleep
import os
import numpy as np
from sklearn.metrics import classification_report
```

### User Login 

```python
status, msg = ds.auth_login_user('henryvuong2408@yahoo.com','j2ZZKMwMer')
if not status:
    print(msg)
```

### Data Upload and Clean

```python
# Preview dataset
df = pd.read_csv(os.path.join(PATH_TO_DATASET, TRAIN_DATASET))
df.head()
```

**Upload dataset to Darwin**

```python
# Upload dataset
status, dataset = ds.upload_dataset(os.path.join(PATH_TO_DATASET, TRAIN_DATASET))
if not status:
    print(dataset)
```

**Clean dataset**

```python
# clean dataset
target = "Wins"
status, job_id = ds.clean_data(TRAIN_DATASET, target = target)

if status:
    ds.wait_for_job(job_id['job_name'])
else:
    print(job_id)
```

### Create and Train Model 

```python
model = target + "_model0" + ts
ds.delete_model(model)
status, job_id = ds.create_model(dataset_names = TRAIN_DATASET, \
                                 model_name =  model, \
                                 max_train_time = '00:02')
if status:
    ds.wait_for_job(job_id['job_name'])
else:
    print(job_id)
```

### Extra Training

```python
# Train some more
status, job_id = ds.resume_training_model(dataset_names = TRAIN_DATASET,
                                          model_name = model,
                                          max_train_time = '00:05')
                                          
if status:
    ds.wait_for_job(job_id['job_name'])
else:
    print(job_id)
```

### Analyze Model

```python
# Retrieve feature importance of built model
status, artifact = ds.analyze_model(model)
sleep(1)
if status:
    ds.wait_for_job(artifact['job_name'])
else:
    print(artifact)
status, feature_importance = ds.download_artifact(artifact['artifact_name'])
```

Show the 10 most important features of the model.

```python
feature_importance[:10]
```

## Predictions
**Perform model prediction on the the training dataset.**

```python
status, artifact = ds.run_model(TRAIN_DATASET, model)
sleep(1)
ds.wait_for_job(artifact['job_name'])
```

Download predictions from Darwin's server.

```python
status, prediction = ds.download_artifact(artifact['artifact_name'])
prediction.head()
```

After running our training dataset to our model, we produced a 82% weigthed accuracy. The precision and recall is shown below.

```python
unq = df[target].unique()[::-1]
p = np.zeros((len(prediction),))
a = np.zeros((len(prediction),))
for i,q in enumerate(unq):
    p += i*(prediction[target] == q).values
    a += i*(df[target] == q).values
#Plot predictions vs actual
plt.plot(a)
plt.plot(p)
plt.legend(['Actual','Predicted'])
plt.yticks([i for i in range(len(unq))],[q for q in unq]);
print(classification_report(df[target][:239], prediction[target][:239]))
```

**Perform model prediction on a test dataset that wasn't used in training.** <br>
**Upload test dataset**

```python
status, dataset = ds.upload_dataset(os.path.join(PATH_TO_DATASET, TEST_DATASET))
if not status:
    print(dataset)
```

**Clean test dataset**

```python
# clean test dataset
status, job_id = ds.clean_data(TEST_DATASET, target = target, model_name = model)

if status:
    ds.wait_for_job(job_id['job_name'])
else:
    print(job_id)
```

**Run model on test dataset.**

```python
status, artifact = ds.run_model(TEST_DATASET, model)
sleep(1)
ds.wait_for_job(artifact['job_name'])
```

After running the test dataset to our model, our accuracy only dropped slightly to an 82% weighted finish. This is much better than our Darwin Model v1.0 as that only produced a 60% weighted accuracy.

```python
# Create plots comparing predictions with actual target
status, prediction = ds.download_artifact(artifact['artifact_name'])
df = pd.read_csv(os.path.join(PATH_TO_DATASET, TEST_DATASET))
unq = prediction[target].unique()[::-1]
p = np.zeros((len(prediction),))
a = np.zeros((len(prediction),))
for i,q in enumerate(unq):
    p += i*(prediction[target] == q).values
    a += i*(df[target] == q).values
#Plot predictions vs actual
plt.plot(a)
plt.plot(p)
plt.legend(['Actual','Predicted'])
plt.yticks([i for i in range(len(unq))],[q for q in unq]);
print(classification_report(df[target][:60], prediction[target][:60]))
```

## Using our model to predict the 2019 Playoffs:

Now that we hav our model trained, let's upload the 2019 NBA season and see which teams were overvalued and undervalued

```python
# upload nba 2019 dataset
nba_2019_DATASET = 'nba_2019.csv'
status, dataset = ds.upload_dataset(os.path.join(PATH_TO_DATASET, nba_2019_DATASET))
# clearn dataset
status, job_id = ds.clean_data(nba_2019_DATASET, target = target, model_name = 'Wins_model020190421134148')
# run model on dataset
status, artifact = ds.run_model(nba_2019_DATASET, 'Wins_model020190421134148')
sleep(1)
ds.wait_for_job(artifact['job_name'])
```

```python
# Create plots comparing predictions with actual target
status, prediction = ds.download_artifact(artifact['artifact_name'])
df = pd.read_csv(os.path.join(PATH_TO_DATASET, nba_2019_DATASET))
unq = prediction[target].unique()[::-1]
p = np.zeros((len(prediction),))
a = np.zeros((len(prediction),))
for i,q in enumerate(unq):
    p += i*(prediction[target] == q).values
    a += i*(df[target] == q).values
#Plot predictions vs actual
plt.plot(a)
plt.plot(p)
plt.legend(['Actual','Predicted'])
plt.yticks([i for i in range(len(unq))],[q for q in unq]);
print(classification_report(df[target][:30], prediction[target][:30]))
```

There were a total of eight teams total that over/under performed expectation, four of those teams being playoff contenders. The Denver Nuggets, LA Clippers, and Houston Rockets won more games than expected and the Utah Jazz won less games than expected.

Therefore, one might assume that the series between the Rockets and Jazz (currently going on at the time of this paper, with the Rockets leading 3-0) would be closer than expected. However, the Rockets lead the Jazz 3-0 in the series. Although the Rockets won more games than expected, they are still a very good team (this roster did make the Western Conference Finals last year and take the champion Warriors to 7 games), and so maybe an undervalued Rockets team is still an incredibly talented roster. The Denver Nuggets are currently playing the San Antonio Spurs, and as a 2-seed the Nuggets are heavily favored. But our model shows the Nuggets are overvalued, and the series is currently tied at 2 games each. Our predictions are currently a mixed bag, with some being more accurate than others, but given more data, we are sure our model will be more accurate.
